# **Outliers**

Para detectar y eliminar los outliers usare la tecnica de 3 sigmas debido a que mos permite conservar una mayor mas de los datos. Con ellos busco eliminar aquellos outliers 'malos' que afectarian de forma negativa nuestro modelo. Recordando que para resolver nuestro problema debemos usar modelos de regresion y estos son muy sensibles a los valores atipicos.

**3 SIGMAS**

la regla de los 3 sigmas basa en definir limites superior e inferior a los datos, a partir de la siguiente formula:

Limite inferior: u - (3 * s).

Limite superiro: u + (3 * s).

Donde u es el promedio de los datos y s la desviacion estandar. Es decir tomaremos los datos que esten ubicados 3 desviaciones estandar por debajo del promedio y 3 desviaciones estandar por encima del promedio.

````python
def aplicar_3_sigmas(col):

  valor_medio = col.mean()
  std = col.std()
  minimo = valor_medio - 3*std
  maximo = valor_medio + 3*std
  print('Valor Minimo: ',minimo)
  print('Valor Maximo: ',maximo)
  mascara_outliers = np.logical_or(col< minimo, col > maximo)

  return mascara_outliers
````

# Imputacion de Valores Faltantes

Como hemos leido y aprendido de en las clases, eliminar las instancias que poseen valores faltantes, no siempre es una buena estrategia debido a que al eliminar dichas instancias, se pierde informacion que nuestro modelo podria aprovechar. Para ello se creo una estrategia de imputacion.

Para este caso usare KNN Imputer debido a que me parecio la estegia mas correcta, ya que usa todas las columnas del dataframe para imputar y usa K-vecinos mas cercanos.

KNN Imputer

Al instanciar el KNNImputer le asignamos uno valores para algunos hiperparametros:

missing_values: Le decimos que tipo de formato tendran los valores nulos o vacios que tenga nuestra matriz.

n_neighbors: numero de vecinos que tendra en cuenta a la hora de promediar y llenar los valores faltantes.

add_indicator: agregar una columna donde indicara a que instacias se les imputo algun valor.

````python
from sklearn.impute import KNNImputer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
rmse = lambda y, yhat: np.sqrt(mean_squared_error(y, yhat))
````

```
# Tiene formato de código
%%time
errors = []
for k in range(1, 20, 1):
    knn_imputer = KNNImputer(n_neighbors=k, add_indicator=True)
    X_train_imp = knn_imputer.fit_transform(X_train)
    X_test_imp = knn_imputer.transform(X_test)
    
    model = LinearRegression()
    model.fit(X_train_imp, y_train)
    preds = model.predict(X_test_imp)
    error = rmse(y_test, preds)
    errors.append({'K': k, 'RMSE': error})
```
